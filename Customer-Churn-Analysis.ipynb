# Module 2: Explatory Data Analysis 

```python
# Converting the TotalCharges into float as well as replacing the blank with o as tenure is 0
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")

df.info()
```

```python
# Checking the missing values in dataset

print("Missing value?", df.isnull().values.any())
```

```python
df.isnull().sum()
```

```python
# Check for the Duplicate Data
df.duplicated().any() # Since the data does not contain any duplicate value
```

```python
# To drop the duplicate data value 
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
#Converting the value in SeniorCitizen 
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

# Module 3:  Data Visualization

```python
# Display the number of Churn by customer
df.columns
```

```python
df['Churn'].value_counts()
```

```python
fig = plt.figure(figsize=(9, 6))
sns.countplot(x='Churn', data=df, hue="Churn")
plt.title("Number of Customers by Churn")
plt.tight_layout
plt.show()
```

```python
fig = plt.figure(figsize=(9, 6))
ax = sns.countplot(x='gender', data=df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Gender")
plt.tight_layout
plt.show()
```

# The Churn rate of male and female is similar as per according to the above graph 

```python
# Analyze numerical variables
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.tight_layout()
    plt.show()
```

```python
# Convert 'Churn' column to numerical values
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Select numerical columns
numerical_cols = df.select_dtypes(include='number').columns.tolist()

# Analyze relationship between numerical variables and churn
plt.figure(figsize=(12, 8))
sns.heatmap(df[numerical_cols + ['Churn']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
